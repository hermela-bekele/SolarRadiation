```python
import pandas as pd

# Load dataset
data = pd.read_csv('benin-malanville.csv')

# Display the first few rows
data.head()
```

```python
# check the shape of the dataset
print(f"Rows: {data.shape[0]}, Columns: {data.shape[1]}")

# display column names
print(f"Columns: {data.columns.tolist()}")

# get data types and memory usage
data.info()
```

```python
# display summary statistics for numerical columns
data.describe()
```

```python
# check for missing values and display counts per column
data.isnull().sum()
```

```python
# check for duplicate rows
print(f"Duplicate rows: {data.duplicated().sum()}")

# remove duplicate rows
data = data.drop_duplicates()
```

```python
# handling missing values drop missing rows
data = data.dropna()
```

```python
# using boxplot to visually display data distribution and highlights outliers
import seaborn as sns
import pandas as pd
data = pd.read_csv('benin-malanville.csv')
sns.boxplot(data['GHI'])
```

```python
sns.scatterplot(x='ModA', y='ModB', data=data)
```

```python
# calculate z-score for one colum
from scipy.stats import zscore
data['zscore'] = zscore(data['Precipitation'])
outliers = data[data['zscore'].abs() > 3]
```

```python
from scipy.stats import zscore

# claculate z-score for all numerical colums
z_scores = data.select_dtypes(include=['float64', 'int64']).apply(zscore)

# find outliers where zscore > 3 or zscore < 3
outliers = (z_scores.abs() > 3)

# list outliers indices for each column
outlier_indices = {column: data[outliers[column]].index.tolist() for column in z_scores.columns}
print(outlier_indices)
```

```python
# visualizing trends over time
data['GHI'].plot(figsize=(10, 6), title = 'GHI over time')
```

```python
# visual correlation using heatmap for better understanding
import seaborn as sns
import matplotlib.pyplot as plt

numeric_data = data.select_dtypes(include=['float64', 'int64'])
correlation_matrix = numeric_data.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
